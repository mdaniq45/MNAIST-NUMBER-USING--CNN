```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

# 1. Data Preparation
transform = transforms.ToTensor()
dataset = datasets.MNIST(root="data", train=True, transform=transform, download=True)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# 2. Define VAE Model
class VAE(nn.Module):
    def __init__(self, input_dim=784, hidden_dim=400, latent_dim=20):
        super(VAE, self).__init__()
        # Encoder
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)
        # Decoder
        self.fc3 = nn.Linear(latent_dim, hidden_dim)
        self.fc4 = nn.Linear(hidden_dim, input_dim)

    def encode(self, x):
        h = torch.relu(self.fc1(x))
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h = torch.relu(self.fc3(z))
        return torch.sigmoid(self.fc4(h))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

# 3. Loss Function
def loss_function(recon_x, x, mu, logvar):
    BCE = nn.functional.binary_cross_entropy(recon_x, x, reduction='sum')
    KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return BCE + KLD

# 4. Training the VAE
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vae = VAE().to(device)
optimizer = torch.optim.Adam(vae.parameters(), lr=1e-3)

for epoch in range(10):
    for batch in dataloader:
        x, _ = batch
        x = x.view(-1, 784).to(device)
        optimizer.zero_grad()
        recon_x, mu, logvar = vae(x)
        loss = loss_function(recon_x, x, mu, logvar)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item()}")

# 5. Generating New Data
vae.eval()
z = torch.randn(1, 20).to(device)  # Random latent vector
generated = vae.decode(z).view(28, 28).cpu().detach().numpy()

import matplotlib.pyplot as plt
plt.imshow(generated, cmap="gray")
plt.title("Generated Image")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# Set device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Define the Generator
class Generator(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Generator, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, output_dim),
        )

    def forward(self, x):
        return self.net(x)

# Define the Discriminator
class Discriminator(nn.Module):
    def __init__(self, input_dim):
        super(Discriminator, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 1),
            nn.Sigmoid(),
        )

    def forward(self, x):
        return self.net(x)

# Initialize the Generator and Discriminator
input_dim = 2  # Latent space size
data_dim = 2   # Real data dimensions
generator = Generator(input_dim, data_dim).to(device)
discriminator = Discriminator(data_dim).to(device)

# Optimizers
lr = 0.0002
optimizer_G = optim.Adam(generator.parameters(), lr=lr)
optimizer_D = optim.Adam(discriminator.parameters(), lr=lr)

# Loss function
criterion = nn.BCELoss()

# Training Parameters
epochs = 5000
batch_size = 64
latent_dim = input_dim

# Generate real data (2D Gaussian distribution)
def real_data_sampler(n):
    return torch.randn(n, data_dim).to(device)

# Train GAN
for epoch in range(epochs):
    # Step 1: Train Discriminator
    real_data = real_data_sampler(batch_size)
    fake_data = generator(torch.randn(batch_size, latent_dim).to(device)).detach()
    real_labels = torch.ones(batch_size, 1).to(device)
    fake_labels = torch.zeros(batch_size, 1).to(device)

    optimizer_D.zero_grad()
    real_loss = criterion(discriminator(real_data), real_labels)
    fake_loss = criterion(discriminator(fake_data), fake_labels)
    d_loss = real_loss + fake_loss
    d_loss.backward()
    optimizer_D.step()

    # Step 2: Train Generator
    noise = torch.randn(batch_size, latent_dim).to(device)
    fake_data = generator(noise)
    fake_labels = torch.ones(batch_size, 1).to(device)

    optimizer_G.zero_grad()
    g_loss = criterion(discriminator(fake_data), fake_labels)
    g_loss.backward()
    optimizer_G.step()

    # Print progress
    if epoch % 500 == 0 or epoch == epochs - 1:
        print(f"Epoch [{epoch}/{epochs}] | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

# Generate and plot synthetic data
with torch.no_grad():
    noise = torch.randn(1000, latent_dim).to(device)
    generated_data = generator(noise).cpu().numpy()

plt.scatter(generated_data[:, 0], generated_data[:, 1], alpha=0.5, label="Generated Data")
real_samples = real_data_sampler(1000).cpu().numpy()
plt.scatter(real_samples[:, 0], real_samples[:, 1], alpha=0.5, label="Real Data")
plt.legend()
plt.title("Real vs Generated Data")
plt.show()
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import Trainer, TrainingArguments

# Load a pre-trained BERT model and tokenizer
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Example data
texts = ["I love machine learning!", "This is a bad day."]
labels = [1, 0]  # 1: positive, 0: negative

# Tokenize input data
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    evaluation_strategy="steps",
)

# Define a Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=inputs,
    eval_dataset=inputs,
)

# Train the model
trainer.train()
```
